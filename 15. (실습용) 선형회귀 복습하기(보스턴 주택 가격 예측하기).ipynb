```python
import seaborn as sns
plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(data=data_f, annot=True, fmt='.2f', linewidth=.5, cmap='RdYlGn_r')
plt.show()
```

```python
#회귀 직선 그리는 그래프
fig, axs = plt.subplots(figsize=(8,8), ncols=2, nrows=2)

lm_features = ['LSTAT','RM','PTRATIO','INDUS']

for i, feature in enumerate(lm_features):
    row = int(i/2) # 2행
    col = i%2

    # sns.regplot : 회귀직선을 그려줌
    sns.regplot(x=feature, y='MEDV', data=data, ax=axs[row][col])
```

### 집값 상승과 상관관계가 가장 높은 것 : RM
