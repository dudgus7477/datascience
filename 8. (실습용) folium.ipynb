### 여러위치에 마커 표시하기

#### 마커에 텍스트 표시하기 - 툴팁으로 표현

```
folium.Marker([마커 표시 위도, 마커 표시 경도], tooltip='텍스트').add_to(지도이름)
```


#### 마커에 텍스트 표시하기 - 팝업으로 표현
```
folium.Marker([마커 표시 위도, 마커 표시 경도], popup='텍스트').add_to(지도이름)
```
